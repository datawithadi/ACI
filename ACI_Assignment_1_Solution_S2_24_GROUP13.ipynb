```python
#Code Block : Set Initial State (Must handle dynamic inputs)

# Take input from the user
from prettytable import PrettyTable

def get_knapsack_items():
    num_items = int(input("Enter Total Number of Items: "))

    weights = []
    values = []
    ratios = []

    print("\nEnter weights and values for each item:")
    for i in range(num_items):
        w = float(input(f"  Item {i+1} - Weight: "))
        v = float(input(f"          - Value : "))
        weights.append(w)
        values.append(v)
        ratios.append(round(v / w, 2))

    # Display the data as a table
    table = PrettyTable()
    table.field_names = ["Item", "Weight", "Value", "Value/Weight Ratio"]
    for i in range(num_items):
        table.add_row([f"Item {i+1}", weights[i], values[i], ratios[i]])

    print("\nItem Data Table")
    print(table)

    list_items = list(zip(weights, values))

    return list_items, num_items, weights, values, ratios
```

For Transition and Cost Matrix --> Convert Knapsack Data into Structure Suitable for Ant Colony Optimization.

**Transition Matrix (τ)** — stores the pheromone levels between item selections (initially small values, updated over time).

**Heuristic Information (η)**— based on value-to-weight ratio(more desirable items should be chosen earlier).

**Cost Matrix (optional)** — since this is not a path-based TSP problem, the "cost" is indirectly defined by weight/value constraints.

```python
## Matrix Function
import numpy as np
import pandas as pd
from IPython.display import display

def initialize_matrices(num_items, ratios):
    # Initialize pheromone matrix (τ) - symmetric with small initial pheromone
    tau = np.full((num_items, num_items), 0.1)

    # Heuristic matrix (η) - use ratio of value/weight as desirability
    eta = np.zeros((num_items, num_items))
    for i in range(num_items):
        for j in range(num_items):
            if i != j:
                eta[i][j] = ratios[j]  # desirability of going from i to j

    # Convert to pandas DataFrame for display
    tau_df = pd.DataFrame(tau, columns=[f"Item {i+1}" for i in range(num_items)],
                               index=[f"Item {i+1}" for i in range(num_items)])

    eta_df = pd.DataFrame(eta, columns=[f"Item {i+1}" for i in range(num_items)],
                               index=[f"Item {i+1}" for i in range(num_items)])

    print("Initial Pheromone Matrix (τ):")
    display(tau_df)

    print("\nHeuristic Matrix (η) based on Value/Weight Ratio:")
    display(eta_df)

    return tau, eta
```

```python
#Code block : Write fucntion to handle goal test (Must handle dynamic inputs). Ideally this would be called while search algorithms are implemented

def get_aco_parameters():
    alpha = float(input("Enter Pheromone Influence (α): "))
    beta = float(input("Enter Heuristic Weight (β): "))
    iterations= int(input("Enter Number of Iterations: "))
    max_capacity = float(input("Enter Maximum Capacity of the Knapsack (in kg): "))
    no_of_ants = int(input("Enter Number of Ants: "))

    return alpha, beta, iterations, max_capacity, no_of_ants
```

```python
# Utility Function Print ACO Result

def print_aco_solution(best_solution, best_value):
    result_table = PrettyTable()
    result_table.field_names = ["Selected Item", "Weight", "Value"]

    total_weight_result = 0
    total_value_result = 0

    for item_index in best_solution:
        weight = list_items[item_index][0]
        value = list_items[item_index][1]
        result_table.add_row([f"Item {item_index + 1}", weight, value])
        total_weight_result += weight
        total_value_result += value

    print("\nACO Result - Selected Items:")
    print(result_table)

    print(f"\nFinal Result:")
    print(f"Total Selected Weight: {total_weight_result:.2f}")
    print(f"Total Selected Value: {total_value_result:.2f}")
```

```python
# Utility Function Print Final Pheromones Result

def print_final_pheromones(pheromones):
    pheromone_table = PrettyTable()
    pheromone_table.field_names = ["Item", "Pheromone Level"]

    # Assuming pheromones is a 1D array where index corresponds to item index
    for i in range(len(pheromones)):
      pheromone_table.add_row([f"Item {i+1}", f"{pheromones[i]:.4f}"])

    print("\nFinal Pheromone Levels:")
    print(pheromone_table)

```

### 2.	Definition of Algorithm  (Ant Colony Optimization)

```python
#Code Block : Function for algorithm implementation

def ant_colony_optimization(list_items, alpha, beta, iterations, ratios, max_capacity, no_of_ants):

    num_items = len(list_items)
    pheromones = np.ones(num_items)
    best_value = 0
    best_solution = []

    heuristics = [value / weight for weight, value in list_items]

    for _ in range(iterations):
        all_solutions = []
        for _ in range(no_of_ants):
            solution = []
            total_weight = 0
            available_items = list(range(num_items))

            while available_items:
                probabilities = []
                for i in available_items:
                    if total_weight + list_items[i][0] <= max_capacity:
                        tau = pheromones[i] ** alpha
                        eta = heuristics[i] ** beta
                        probabilities.append(tau * eta)
                    else:
                        probabilities.append(0)

                if sum(probabilities) == 0:
                    break

                probabilities = [p / sum(probabilities) for p in probabilities]
                chosen = np.random.choice(available_items, p=probabilities)
                solution.append(chosen)
                total_weight += list_items[chosen][0]
                available_items.remove(chosen)

            total_value = sum(list_items[i][1] for i in solution)
            if total_value > best_value:
                best_value = total_value
                best_solution = solution

            all_solutions.append((solution, total_value))

        # Update pheromones (evaporation and reinforcement)
        pheromones *= 0.9  # evaporation
        for solution, value in all_solutions:
            for i in solution:
                pheromones[i] += value / 100.0



    return best_solution, best_value, pheromones
```

<h1>#Code Block : Print the Time & Space complexity of algorithm</h1>

<h2>Time Complexity Analysis</h2>

<p><b>Let:</b></p>
<ul>
  <li><b>n</b> be the number of items.</li>
  <li><b>m</b> be the number of ants.</li>
  <li><b>iter</b> be the number of iterations.</li>
</ul>

<h3>1. Initialization</h3>
<ul>
  <li><b>Taking input of n items:</b><br><b>Time Complexity:</b> O(n)</li>
  <li><b>Building ratio list:</b><br><b>Time Complexity:</b> O(n)</li>
  <li><b>Initializing τ and η matrices of size n * n:</b><br><b>Time Complexity:</b> O(n<sup>2</sup>)</li>
</ul>

<h3>2. Main ACO Loop</h3>
<p>Each ant builds a solution per iteration.</p>
<p>For each iteration (total <i>iter</i>):</p>
<ul>
  <li><b>For each ant (m):</b></li>
  <ul>
    <li>Builds solution by evaluating each of n items:<br>
        <b>Time Complexity:</b> O(n<sup>2</sup>) <i>(since we check transitions between pairs)</i></li>
    <li>Local fitness evaluation and weight check:<br>
        <b>Time Complexity:</b> O(n)</li>
  </ul>
</ul>
<p><b>So,</b></p>
<ul>
  <li><b>Total Time per Iteration:</b> O(m x n<sup>2</sup>)</li>
  <li><b>Total Time Overall:</b> O(iter x m x n<sup>2</sup>)</li>
</ul>

<h3>3. Pheromone Update</h3>
<ul>
  <li>After each iteration:<br>
      <b>Time Complexity:</b> O(n<sup>2</sup>)</li>
</ul>

<hr>

<h2>Space Complexity Analysis</h2>

<h3>1. Input Storage:</h3>
<ul>
  <li><code>weights</code>, <code>values</code>, <code>ratios</code>: <b>Space Complexity:</b> O(n)</li>
</ul>

<h3>2. Matrices:</h3>
<ul>
  <li><code>τ</code> and <code>η</code> of size n * n:<br>
      <b>Space Complexity:</b> O(n<sup>2</sup>)</li>
</ul>

<h3>3. Solution Construction:</h3>
<ul>
  <li>Temporary arrays per ant: <b>O(n)</b></li>
  <li>Best solution list: <b>O(n)</b></li>
</ul>

<h3><b>So,</b></h3>
<ul>
  <li><b>Total Space Complexity:</b> O(n<sup>2</sup>)</li>
</ul>
